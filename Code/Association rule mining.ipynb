```python
import pandas as pd
import numpy as np
import mlxtend
from mlxtend.frequent_patterns import apriori, association_rules
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../Data/raw_data.csv',engine='python')
df = df.drop(columns='Season')
df = df.drop(columns='CreatedOn')
df.head()
```

```python
for column in df[['Sector', 'Category','DistrictName','BlockName','Crop', 'QueryType', 'StateName']]:
    cnt=0
    for row in df[column]:
        try:
            int(row)
            df.loc[cnt, column]=np.nan
        except ValueError:
            pass
        cnt+=1
```

```python
print (df.isnull().sum())
df=df.dropna()
df.head()
```

```python
df.reindex()
```

```python
df = df.reset_index()
```

```python
df.head()
```

```python
df=df.drop(columns='index')
```

```python
df.head()
```

```python
df_onehot = pd.get_dummies(df,prefix=['Sector','Category','Crop','QueryType','StateName','DistrictName','BlockName'])
```

```python
df_onehot.head()
```

```python
#df_onehot.to_csv('../Data/onehot_season_and_time_removed.csv', encoding='utf-8-sig', index=False)
```

```python
freq_items = apriori(df_onehot, min_support=0.2, use_colnames=True, verbose=1)
freq_items.head(7)
```

```python
rules = association_rules(freq_items, metric="confidence", min_threshold=0.6)
rules.head()
```

```python
#rules.to_csv('../Data/Association_rules_obtained_with_others_value.csv', encoding='utf-8-sig', index=False)
```
